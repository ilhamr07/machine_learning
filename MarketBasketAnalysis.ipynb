```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#impor data transaksi ke data frame

df = pd.read_csv('Salinan Salinan Salinan Online Retail Data.csv')
```

```python
#cleansing data 
df_clean = df.copy()
df_clean = df_clean[df_clean['order_id'].str[:1]!='C']
```

```python
#membuat pivot tabel berdasarkan product code
basket = pd.pivot_table(df_clean,index='order_id',columns='product_name',values='product_code',aggfunc='nunique',fill_value=0)
basket
```

```python
basket.shape
```

```python
#encode

def encode(x):
    if x ==0:
        return False
    if x>=0:
        return True
basket_encode = basket.applymap(encode)
basket_encode
```

```python
#filter ambil transaksi dengan banyak produk lebih dari 1

filter_basket = basket_encode[(basket_encode>0).sum(axis=1)>1]
```
