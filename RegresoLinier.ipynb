```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Generate Data

```python
jumlah_data= 100

# hasil=[]

# for a in range(jumlah_data):
#     hasil.append(a + np.random.rand())
# print(hasil)


y = np.array([i*0.1 + np.random.randn() for i in range(jumlah_data)])

x = np.array([i*0.1 for i in range(jumlah_data)])
```

Plot Data

```python
#membuat plot
plt.scatter(x,y)

plt.axis([0,10,0,10])

#menampilkan plot
#plt.show()
```

Fungsi Linear

```python
def regresi_linear(x,gradien):
    y = gradien*x
    return y
```

Menghitung persamaan linear awal

```python
x_prediction = np.array([-1,11,1])#awal, akhir, range
m_prediction = 1
y_prediction = regresi_linear(x_prediction, m_prediction)
```

```python
plt.scatter(x,y)
plt.plot(x_prediction, y_prediction,color='r')
plt.show()
```
